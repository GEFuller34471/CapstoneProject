```python
import pandas as pd
import calendar
import datetime
from datetime import datetime
from datetime import timedelta
from datetime import date
from dateutil.relativedelta import relativedelta

date_year = pd.read_csv('olist_date_year.csv')

# set Date to date datatype
date_year['Date'] = date_year['Date'].apply(pd.to_datetime)

# add month and dayone to calculate fdom - first_day_of_month
date_year['month'] = pd.DatetimeIndex(date_year['Date']).month
##date_year['dayone']  = 1

#Add day of month number 
date_year['dom#'] = pd.DatetimeIndex(date_year['Date']).day

#Add Day of week column
def get_dow_number(index):
    if index == 6:
        return 1
    else:  
        return index+2


date_year['dow#'] = pd.DatetimeIndex(date_year['Date']).weekday
date_year['dow#'] = date_year['dow#'].apply(get_dow_number)

#Add month name column
month_names =["January","February","March","April","May","June","July","August","September","October","November","December"]
def get_monthname(index):
    return month_names[index-1]
date_year['month_name'] = date_year['month'].apply(get_monthname)

#Add month short name column
month_abb=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
def get_monthabb(index):
    return month_abb[index-1]
date_year['month_short_name'] = date_year['month'].apply(get_monthabb)

#Add Dayname column
day_name= ["Sunday","Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday","Sunday",]
def get_dayname(index):
    return day_name[index-1]
date_year['dow_name'] = date_year['dow#'].apply(get_dayname)


#Concat year, month and day to string with two digits for month & day
#date_year['ymd'] = date_year['DateYear'].astype(str) + date_year['month'].astype(str).str.zfill(2) +  date_year['dayone'].astype(str).str.zfill(2)


#Calulate the first day of month
#date_year['fdom'] = pd.to_datetime(date_year['ymd'], format='%Y%m%d')
#delete unneeeded columns
#del date_year['ymd']
#del date_year['month'] 
#del date_year['dayone']


#Calculate the last day of month by subtracting one day from first day of month and then add one month
##date_year['ldom'] = date_year['fdom'] - timedelta(days = 1)
##date_year['ldom'] = date_year['ldom'] + pd.DateOffset(months=1)

print(date_year.info())
print(date_year.head(33))
```
